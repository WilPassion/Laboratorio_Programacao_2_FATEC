```python
# Cifra de CÃ©sar (ROT13)

# criando lista alfabeto
lst_alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
                'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# Gerar lista alfabeto deslocada 13 posiÃ§Ãµes - ROT13
rot13 = []
for letra in lst_alfabeto:
    rot13.append(lst_alfabeto[(lst_alfabeto.index(letra) + 13) % 26])  # Desloca cada letra 13 posiÃ§Ãµes

# Solicita ao usuÃ¡rio a opÃ§Ã£o de criptografar ou decriptar
acao = input("VocÃª deseja (c)riptografar ou (d)ecriptar uma mensagem? ").lower()

# Solicita a mensagem do usuÃ¡rio
mensagem = input("Digite a mensagem: ").lower()

# Processa a mensagem com base na aÃ§Ã£o escolhida
resultado = []
for letra in mensagem:
    if letra in lst_alfabeto:
        # Encontra a letra correspondente na lista rot13
        indice = lst_alfabeto.index(letra)
        resultado.append(rot13[indice])
    else:
        # MantÃ©m caracteres que nÃ£o estÃ£o no alfabeto
        resultado.append(letra)

# Exibe a mensagem resultante
print("Resultado:", ''.join(resultado))
```

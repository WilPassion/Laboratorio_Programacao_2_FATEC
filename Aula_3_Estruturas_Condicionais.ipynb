```python
# EXERCÍCIO PRÁTICO 1
# Elabore um algoritmo que receba duas notas e calcule a média final. Se a média final for
# maior igual a 6,0, mostre a mensagem “Você está aprovado!”

# Entrada:
flt_nota1 = float(input('Digite a primeira nota: '))
flt_nota2 = float(input('Digite a segunda nota: '))

# Processamento:
media = (flt_nota1 + flt_nota2) / 2

# Saída:
if media >= 6:
  print('Você está aprovado! =D')
```

```python
# Atividade 2 – exercício B

# Entrada dados - salário bruto do funcionário (em Reais)
flt_salario = float(input('Digite o seu salário: R$'))
int_dependentes = int(input('Digite a quantidade de dependentes maiores de 21 anos: '))

# Função cálculo: contribuição INSS 2023
def calcula_INSS(salario):
    if salario <= 1302:
        deducao_INSS = salario * 0.075
    elif salario <= 2571.29:
        deducao_INSS = salario * 0.09
    elif salario <= 3856.94:
        deducao_INSS = salario * 0.12
    elif salario <= 7507.49:
        deducao_INSS = salario * 0.14
    else:
        deducao_INSS = 0.14 * 7507.49  # valor fixo para salários acima de R$ 7.507,49
    return deducao_INSS

# Função cálculo: contribuição IRRF 2023
def calcula_IRRF(salario):
    deducao_dependentes = int_dependentes * 189.59
    salario_base = salario - deducao_dependentes  # Salário base após dedução dos dependentes

    if salario_base <= 2112:
        deducao_IRRF = 0
    elif salario_base <= 2826.65:
        deducao_IRRF = (salario_base * 0.075) - 158.40
    elif salario_base <= 3751.05:
        deducao_IRRF = (salario_base * 0.15) - 370.40
    elif salario_base <= 4664.68:
        deducao_IRRF = (salario_base * 0.225) - 651.73
    else:
        deducao_IRRF = (salario_base * 0.275) - 884.96
    return deducao_IRRF

# Função para formatar valores em moeda brasileira
def formata_valor(valor):
    return f"R$ {valor:_.2f}".replace('.', ',').replace('_', '.')

# Processamento - Calcular descontos
deducao_INSS = calcula_INSS(flt_salario)  # Calcula a dedução do INSS
salario_apos_INSS = flt_salario - deducao_INSS  # Salário após desconto do INSS
deducao_IRRF = calcula_IRRF(salario_apos_INSS)  # Calcula a dedução do IRRF
salario_liquido = salario_apos_INSS - deducao_IRRF  # Salário líquido final

# Saída dos dados
print('\nSalário Bruto: {}\nDesconto INSS: {}\nDesconto IRRF: {}\nSalário Líquido: {}'.format(
    formata_valor(flt_salario),
    formata_valor(deducao_INSS),
    formata_valor(deducao_IRRF),
    formata_valor(salario_liquido)
))
```

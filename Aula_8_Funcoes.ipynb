```python
## Funcao_1

# soma_digito (Passo 1 e Passo 2)
def soma_digitos_cpf(cpf):  # CPF Teste - 00393970841
    lst_digitos = []

    # Armazenando apenas os dígitos numéricos
    for digito in cpf:
        if digito.isdigit():
            lst_digitos.append(int(digito))

    # Soma dos dígitos (Passo 1)
    soma = sum(lst_digitos)

    if soma != 44:
        print("O CPF é inválido")
    else:
        print("Passo 1 validado com sucesso")

        # Passo 2: Cálculo do primeiro dígito verificador
        lst_digitos_verificadores = lst_digitos[:-2]  # Pegando os 9 primeiros dígitos
        verifica_digito = 10  # Começando a multiplicação com 10
        soma_verificadores = 0

        # Loop para multiplicar os dígitos pelos valores decrescentes
        for i in range(9):  # São os 9 primeiros dígitos
            soma_verificadores += lst_digitos_verificadores[i] * verifica_digito
            verifica_digito -= 1

        print(f"Soma dos multiplicadores para o primeiro dígito verificador: {soma_verificadores}")

        # Passo 3: Divisão por 11 e cálculo do primeiro dígito verificador
        resto_divisao = soma_verificadores % 11
        if resto_divisao < 2:
            primeiro_digito_verificador = 0
        else:
            primeiro_digito_verificador = 11 - resto_divisao

        print(f"Primeiro dígito verificador calculado: {primeiro_digito_verificador}")

        # Passo 5: Cálculo do segundo dígito verificador
        lst_digitos_verificadores.append(primeiro_digito_verificador)  # Adiciona o primeiro dígito para o cálculo do segundo
        verifica_digito = 11  # Agora começamos a multiplicar com 11
        soma_verificadores = 0

        # Loop para multiplicar os dígitos pelos valores decrescentes para o segundo dígito
        for i in range(10):  # Agora são os 10 primeiros dígitos
            soma_verificadores += lst_digitos_verificadores[i] * verifica_digito
            verifica_digito -= 1

        print(f"Soma dos multiplicadores para o segundo dígito verificador: {soma_verificadores}")

        # Passo 6: Cálculo do segundo dígito verificador
        resto_divisao = soma_verificadores % 11
        if resto_divisao < 2:
            segundo_digito_verificador = 0
        else:
            segundo_digito_verificador = 11 - resto_divisao

        print(f"Segundo dígito verificador calculado: {segundo_digito_verificador}")

        # Passo 7: Comparação dos dígitos verificadores
        if (primeiro_digito_verificador == lst_digitos[-2]) and (segundo_digito_verificador == lst_digitos[-1]):
            print("O CPF é válido")

            # Passo 8: Verificação da origem do CPF
            verificar_origem_cpf(lst_digitos)
        else:
            print("O CPF é inválido")

## Função para verificar a origem do CPF (Passo 8)
def verificar_origem_cpf(lst_digitos):
    origem_uf = {
        0: "Rio Grande do Sul",
        1: "Distrito Federal, Goiás, Mato Grosso, Mato Grosso do Sul, Tocantins",
        2: "Pará, Amazonas, Acre, Amapá, Rondônia, Roraima",
        3: "Ceará, Maranhão, Piauí",
        4: "Pernambuco, Rio Grande do Norte, Paraíba, Alagoas",
        5: "Bahia, Sergipe",
        6: "Minas Gerais",
        7: "Rio de Janeiro, Espírito Santo",
        8: "São Paulo",
        9: "Paraná, Santa Catarina"
    }

    nono_digito = lst_digitos[8]  # O nono dígito indica a UF de origem
    uf_origem = origem_uf.get(nono_digito, "Origem desconhecida")

    print(f"A origem do CPF é: {uf_origem}")

## Programa Principal
print("-=" * 30)
print("VALIDADOR DE CPF".center(60))
print("-=" * 30)

# Entrada de dados - CPF
int_cpf = input("Digite o número do seu CPF (somente números): ")

# Chamada da função para realizar os passos
soma_digitos_cpf(int_cpf)
```
